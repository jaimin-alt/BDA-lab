```python
import time
import json
import os
import random
from datetime import datetime
from fastavro import writer, parse_schema
```

```python
schema = {
    "doc": "Electricity meter readings",
    "name": "MeterReading",
    "namespace": "example.avro",
    "type": "record",
    "fields": [
        {"name": "meter_id", "type": "string"},
        {"name": "timestamp", "type": "long"},
        {"name": "usage_kwh", "type": "float"}
    ]
}

parsed_schema = parse_schema(schema)
```

```python
def generate_data(n=100000):
    data = []
    for i in range(n):
        record = {
            "meter_id": f"METER_{random.randint(1,1000)}",
            "timestamp": int(datetime.now().timestamp()),
            "usage_kwh": round(random.uniform(0.1, 10.0), 3)
        }
        data.append(record)
    return data
```

```python
def write_avro(data, filename="data.avro"):
    start = time.time()

    with open(filename, "wb") as out:
        writer(out, parsed_schema, data)

    end = time.time()
    return end - start
```

```python
def write_json(data, filename="data.json"):
    start = time.time()

    with open(filename, "w") as out:
        json.dump(data, out)

    end = time.time()
    return end - start
```

```python
data = generate_data(100000)

# Avro
avro_time = write_avro(data)
avro_size = os.path.getsize("data.avro")

# JSON
json_time = write_json(data)
json_size = os.path.getsize("data.json")

# Results
print("\n--- Results ---")
print(f"Avro Size: {avro_size / 1024:.2f} KB")
print(f"JSON Size: {json_size / 1024:.2f} KB")

print(f"Avro Time: {avro_time:.4f} seconds")
print(f"JSON Time: {json_time:.4f} seconds")
```
